## Get Data

```python
import yfinance as yf


data = yf.download("BTC-USD", start="2022-11-01", end="2025-11-01", auto_adjust=False)
data.columns = data.columns.droplevel(1)

#Check the first few rows

data.to_csv("BTC_3years.csv")
data

```
